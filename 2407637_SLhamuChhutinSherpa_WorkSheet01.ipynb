1. Load and Prepare Data:
• Fetch an image of you choice.{If colour convert to grayscale}
• Center the dataset - Standaridze the Data.
• Calculate the covaraince matrix of the Standaridze data.

```python
image1=image_array.copy()
```

```python
display(image1)
```

```python
# 1. Convert to grayscale
gray_img = Image.fromarray(image1).convert('L')
gray_data = np.array(gray_img).astype(np.float64)

# 2. Center the dataset - Standardize the Data
# We subtract the mean from each pixel
mean_val = np.mean(gray_data)
centered_data = gray_data - mean_val
```

```python
# 3. Calculate the covariance matrix
# Typically for images in this context, we treat rows or columns as variables
covariance_matrix = np.cov(centered_data)

print(f"Mean value: {mean_val:.2f}")
print(f"Covariance matrix shape: {covariance_matrix.shape}")
display(covariance_matrix)
```

2. Eigen Decomposition and Identifying Principal Components:
• Compute Eigen Values and Eigen Vectors.
• Sort the eigenvalues in descending order and choose the top k eigenvectors corresponding to
the highest eigenvalues.
• Identify the Principal Components with the help of cumulative Sum plot.

```python
# Compute Eigen Values and Eigen Vectors
eigenvalues, eigenvectors = np.linalg.eig(covariance_matrix)

# Print the shapes to verify results
print(f'Eigenvalues shape: {eigenvalues.shape}')
print(f'Eigenvectors shape: {eigenvectors.shape}')
```

```python
# 1. Get indices that would sort the eigenvalues in ascending order
sorted_indices = np.argsort(eigenvalues)

# 2. Reverse indices for descending order
sorted_indices = sorted_indices[::-1]

# 3. Reorder eigenvalues
sorted_eigenvalues = eigenvalues[sorted_indices]

# 4. Reorder eigenvectors (columns correspond to eigenvalues)
sorted_eigenvectors = eigenvectors[:, sorted_indices]

# 5. Print first few sorted eigenvalues to verify
print('First 5 sorted eigenvalues:', sorted_eigenvalues[:5])
```

```python
# 1. Calculate explained variance ratio
explained_variance_ratio = sorted_eigenvalues / np.sum(sorted_eigenvalues)

# 2. Compute cumulative explained variance
cumulative_variance = np.cumsum(explained_variance_ratio)

# 3. Create the plot
plt.figure(figsize=(10, 6))
plt.step(range(1, len(cumulative_variance) + 1), cumulative_variance, where='mid', label='Cumulative explained variance')
plt.axhline(y=0.95, color='r', linestyle='--', label='95% Explained Variance')
plt.axhline(y=0.90, color='g', linestyle='--', label='90% Explained Variance')

# 4. Add labels and title
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance Ratio')
plt.title('Cumulative Explained Variance vs. Number of Principal Components')
plt.grid(True)
plt.legend(loc='best')

# 5. Identify k for 95% variance
k_95 = np.argmax(cumulative_variance >= 0.95) + 1
print(f'Number of components to explain 95% variance: {k_95}')

plt.show()
```

```python
# Using k=11 as identified in the previous step for 95% variance
k = 11

# 1. Extract the first k columns from the sorted_eigenvectors matrix
top_k_eigenvectors = sorted_eigenvectors[:, :k]

# 2. Verify and print the shape of the resulting array
print(f'Shape of top_k_eigenvectors: {top_k_eigenvectors.shape}')

# Check if it matches expected 159x11
if top_k_eigenvectors.shape == (159, 11):
    print('Successfully extracted top 11 eigenvectors.')
else:
    print(f'Unexpected shape: {top_k_eigenvectors.shape}')
```
